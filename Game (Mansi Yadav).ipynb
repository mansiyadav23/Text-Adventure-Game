```python
################################################################################################
################################################################################################

"""
DocString:

    Introduction:
        In this game you are an Avenger. You can pick if you want to be Iron Man or 
        Captain America. The game is set after the battle of Avengers: Infinity War. 
        An evil Titan named Thanos has wiped out half of the universe and now it is 
        upon the Avengers to undo the snap and bring back the people who were dusted away.

    Round 1: Starting point.
    Round 2.A: Prof Chase.
    Round 2.B: 3 clues. 
    Round 3.A: Captain Marvel 
    Round 3.B: Superman  
    
    Link for Avengers: Infinity war:
        https://www.netflix.com/title/80219127
        
    Extra information: 
        The questions are based on all the movies that are leading up to Infinity War. 

    Known bugs and errors:
        In rounds def as cap_ques and superman_ques:
            For the 3 questions, in an attempt to make the questions random for each time the game is played, 
            I had to make a list of the ques and another list for the answers. 
            In doing so, I had to restrict the user input to '1', '2', and '3'.

            Also, if a person enters 0 or a number greater than 4 for the questions in these 2 rounds they will
            be asked 3 questions again regardless of the questions they may have answered correctly before that.

"""

###############################################################################################
###############################################################################################




import random
import time
import sys
from sys import exit
def game_start():

    global name
    print("""\n
    Welcome Player! \n
    Please pick which Avenger do you wish to be? \n
    1. Captain America 
    2. Iron Man
    """)
    name = input("\nInput your choice: \n").lower().strip()
    if "1" in name or "captain america" in name:
        print("\nHello Cap!")
        name = "Captain America"
        input('\n<Press enter to continue\n>')
        intro()
        
    elif "2" in name or "iron man" in name:
        print("\nHi Tony!")
        name = "Iron Man"
        input('\n<Press enter to continue\n>')
        intro()

    else: 
        print("\nInvalid entry. Please try again. \nPlease enter the number or the name of the Avenger you wish to be.")
        game_start()
    
    
def intro():
    

    print("""
    The Mad Titan Thanos did what he said he would do.
    He wiped out half the universe with a snap using the 
    Infinity Gauntlet. You and the rest of the Avengers 
    failed to stop him.
    
    The world was relying on you and you failed. 
    People lost loved ones. You saw more than half 
    of your team turn to dust. 
    
    Now it\'s upon The Avengers to undo the snap and 
    bring back all the people who were dusted away. 
    The world is in chaos and you all are still trying 
    to figure out where Thanos is. Even if you do find 
    Thanos, you have no idea how to defeat him or undo 
    the snap. 
    """)
    input('\n<Press enter to continue\n>')
    round_1()
    
def round_1():
    print("\nHow you wish to proceed?")
    print("""\n1) Give up \n2) Ask Prof Chase for help \n3) Try to find clues""")
    choice = input(">   \n").lower().strip()
    if "1" in choice or "give up" in choice:
        print(f"\nYou are a coward for giving up {name}!")
        print("Earth was depending on you to save it and you failed.")
        exit(0)
        # If the player gives up at this level, they do not get a chance to try again.
    elif "2" in choice or "ask prof chase for help" in choice:
        print("\nInteresting choice! Let\'s see if Prof Chase can help you.")
        input('\n<Press enter to continue\n>')
        round_2_chase()
    elif "3" in choice or "try to find clues" in choice:
        print("\nYou go searching for clues.")
        input('\n<Press enter to continue\n>')
        round_2_clues()

    else: 
        print("Invalid entry. Please try again.\n")
        round_1() 
        


def round_2_chase():
    print(f"""\n
    Prof Chase: "Welcome {name}!"
    
    {name}: "Hello Professor! We need help to find and defeat Thanos.
    Do you know anything that can help us?"
    
    Prof Chase: "I might be able to help you {name}. 
    But before I help you, I need you to answer a question for me.
    If you can answer it correctly, I will help you. 
    But Beware! 
    If you answer incorrectly, I will trap you in the Quantum Realm for all eternity!"
                
    {name}: "Ummm....
    Okay Professor. What is your question?"
    """)
    # The Quantum Realm is a dimension in the Multiverse.
    # In the Quantum Realm, space and time are believed to be irrelevant.
    input('\n<Press enter to continue\n>')
    print("\nWhat are you allowed to eat in my class?")
    print("\n1) Chips \n2) Carrots \n3) Chocolates")
    choice = input(">   \n").lower().strip()
    if "1" in choice or 'chips' in choice or 'chip' in choice:
        print(f"""\n
    Prof Chase: "Uh Oh! Wrong answer! 
    Off you go to the Quantum Realm where you will be trapped for all eternity."
        """)
        time.sleep(2)
        print("""
    You suddenly see a bright light and Whooosh...
    You are trapped in the Quantum Realm.""")
        fail()
    elif "2" in choice or 'carrots' in choice or 'carrot' in choice:
        print(f"""\n
    Prof Chase: "Right answer! 
    You just saved yourself from spending an eternity in the Quantum Realm."
                
    {name}: "Please help us save the world Professor."
                
    Prof Chase: "Have you heard about Captain Marvel?"
    
    {name}: "No. Who is Captain Marvel?"
    
    Prof Chase: "Captain Marvel is a human-alien hybrid who possess superhuman strength, endurance, 
    and a heightened awareness of danger. She can fly at great speeds, well beyond the sound barrier. 
    She can also shoot photonic light beams from her hands. I believe that she can even time travel."
                
    {name}: "Time travel?"
    
    Prof Chase: "Yes. Time travel. Also, she is strong enough to defeat Thanos.
    I can give you her location but you may need to persuade her to help you."
                
    {name}: "Thank you Professor! The Avengers are forever in your debt.
    If we survive this catastrophe and you ever need our help, we will always be there for you."           
        """) 
        input('\n<Press enter to continue\n>')
        round_3_Captain_Marvel()
    elif "3" in choice or 'chocolates' in choice or 'chocolate' in choice:
        print(f"""\n
    Prof Chase: "Uh Oh! Wrong answer! 
    Off you go to the Quantum Realm where you will be trapped for all eternity."
        """)
        time.sleep(2)
        print("""
    You suddenly see a bright light and Whooosh...
    You are trapped in the Quantum Realm.""")
        fail()

    else: 
        print("\nInvalid entry. Please try again.\n")
        round_2_chase() 
    

def round_2_clues():
    
    print("\nYou find 3 clues. \nWhich one do you wish to use?")
    print("""\n1) Nick Fury \n2) Batman \n3) U.S. Govt""")
    choice = input(">   \n").lower().strip()
    if "1" in choice or 'nick fury' in choice:
        print(f"""\n
    {name}: "Fury! The Avengers need your help! We need a way to find and defeat Thanos."
    
    Fury: "I know someone who can help you. Have you heard about Captain Marvel?"

    {name}: "No. Who is Captain Marvel?"
    
    Fury: "Captain Marvel is a human-alien hybrid who possess superhuman strength, endurance, 
    and a heightened awareness of danger. She can fly at great speeds, well beyond the sound barrier. 
    She can also shoot photonic light beams from her hands. I believe that she can even time travel."
                                
    {name}: "Time travel?"
    
    Fury: "Yes. Time travel. Also, she is strong enough to defeat Thanos.
    I can give you her location but you may need to persuade her to help you."
                
    {name}: "Thanks Fury!"  
    
    Fury: "Go save the world {name}!"
        """)
        input('\n<Press enter to continue\n>')
        round_3_Captain_Marvel()
    elif "2" in choice or 'batman' in choice:
        print(f"""\n
    {name}: "Batman! The Avengers need your help! We need a way to find and defeat Thanos."
    
    Batman: "I know someone who can help you."

    {name}: "Who?"
    
    Batman: "Superman. He is from the planet Krypton. It was destroyed after his birth and which is why he grew up on Earth.
    He possesses the power of flight, superhuman strength, heat vision, cold breath, and super-speed.
    He is strong enough to defeat Thanos."
    
    {name}: "Where can we find him?"  
    
    Batman: "I will tell you where to find him, but convincing him to help you is not going to be easy."
    
    {name}: "Why?"
    
    Batman: "He doesn't trust people easily, especially superheros."
    
    {name}: "I\'m sure we can get him to help us."
        """)
        print("\nBatman tells you where to find Superman.")
        input('\n<Press enter to continue\n>')
        round_3_superman()
    elif "3" in choice or 'us govt' in choice or 'u.s. govt' in choice:
        print("""
    The U.S. Govt blames you for the deaths of all the people who died because of the snap.
    Instead of helping you, the United States Govt puts all the Avengers in an unbreakable prison.
    """)
        fail()

    else: 
        print("\nInvalid entry. Please try again.\n")
        round_2_clues() 

def round_3_Captain_Marvel():
    
    print(f"""
    You manage to find Captain Marvel and tell her what happened with Thanos and how the avengers need her help.
    
    Captain Marvel: "I will help you but I need to confirm if you are really {name} or not. 
    I have been getting reports of an alien species called The Kree that can transform into any being. 
    The Kree are dangerous and I want to be sure that you are not one of them before I help you."
    
    {name}: "How will you be able to determine if I\'m a kree or not by these questions?"
    
    Captain Marvel: "Kree can only access recent memory of the person they turn into. 
    If you can answer these 3 questions correctly I will know if you are Kree or not"
    
    {name}: "Okay. I am ready to answer the questions"
    
    Captain Marvel: "Do remember that if you get even 1 question wrong, I will not hesitate in killing you."
    
    """)
    cap_ques()
    
def cap_ques():
    input("\n<Press enter to continue\n>")
    
    ques_easy_list = ["What weapon does Thor carry?\n1. A bow and arrow\n2. A hammer\n3. A catapult",
                      "Who is the villian in Avengers(2012)?\n1. Vision\n2. Ultron\n3. Loki",
                      "Which of the following avengers can NOT fly?\n1. Vision\n2. Hawkeye\n3. Iron Man",
                      "What is Agent Coulson\'s first name?\n1. Phil\n2. Nick\n3. Steve",
                      "Who is the Vision?\n1. Ultron\n2. Thor\n3. Jarvis",
                      "Where is Black Panther from?\n1. Wakanda\n2. Nakia\n3. T'Chaka",
                      "Where is Thor from?\n1. Midgard\n2. Asgard\n3. Mars",
                      "Who plays Hulk?\n1. Jeremy Renner\n2. Chris Evans\n3. Mark Ruffalo",
                      "Who plays Nick Fury?\n1. Morgan Freeman\n2. Samuel L. Jackson\n3. Denzel Washington",
                      "Who is Loki\'s brother?\n1. Hawkeye\n2. Thor\n3. Hulk"]

    ans_easy_list = ["2",
                     "3",
                     "2",
                     "1",
                     "3",
                     "1",
                     "2",
                     "3",
                     "2",
                     "2"]
    

    counter = 2
    while counter >=0:
        question = random.choice(ques_easy_list)
        print(question)
        
        while True:
            ans = input("\n<Plese enter '1', '2' or '3' only \n>")
            
            try:
                int(ans)
                break
 
            except:
                print("\nYou choose something that is not acceptable. Please choose between '1', '2' and '3'. Choose again.")     
        
        
        # I made people play the game and asked them if they had a problem with answering the questions again
        # if they input 0 or a number bigger than 4. The people who were Marvel fans did not mind answering these 
        # questions again but the people who had not watched Marvel movies and did not know anything about the 
        # Avengers did not enjoy answering the questions again.
        if ans >= '4' or ans == '0':
            print("\nYou choose a wrong number. You have to choose '1', '2' or '3'. \nNow answer these questions again. Be careful while choosing.")
            cap_ques()
        

        elif ans == ans_easy_list[ques_easy_list.index(question)]:
            print("\nYou are right!")
            ans_easy_list.pop(ques_easy_list.index(question))
            ques_easy_list.pop(ques_easy_list.index(question))
            input('\n<Press enter to continue>\n')
            

        elif ans_easy_list.index(ans) != ans_easy_list[ques_easy_list.index(question)]:
            print(f"""
    Captain Marvel: "I was right. You are a Kree."
            """)
            print("\nCaptain Marvel uses her photonic light beams to kill you.")
            ans_easy_list.pop(ques_easy_list.index(question))
            ques_easy_list.pop(ques_easy_list.index(question))
            fail()

        else: 
            print("\nInvalid answer! Please try again")
            cap_ques()
            break

        counter -=1

    cap_win()
        
        
def cap_win():
    print("\nCaptain Marvel believes that you are not a Kree.")
    print("""
    She helps the avengers find and defeat Thanos.
    The snap was undone and all the people who had died because of the snap were saved.
    Congratulations!!
    You won!
    You saved the earth!""")
    exit(0)
    
def round_3_superman():
    
    print("""
    You find Superman but he doesn\'t want to help you. 
    You try to persuade him by explaining that the world needs a hero right now 
    and Superman can be that hero.
    
    Superman agress to help you if only you can answer his 3 questions correctly.
    If you fail to answer even 1 question correctly, Superman will kill you.""")

    superman_ques()
    
def superman_ques():
    input("\n<Press enter to continue\n>")
    
    ques_hard_list = ["What does Stark Industries manufacture in Iron Man 1?\n1. Weapons\n2. Buildings\n3. Chemicals",
                      "Which war did Captain America fight in?\n1. World War I\n2. World War II\n3. Vietnam War",
                      "What is the name of Thor's hammer?\n1. Balder\n2. Mjolnir\n3. Aesir",
                      "What is Peter Quill's alter ego?\n1. Star-Man\n2. Star-Dude\n3. Star-Lord",
                      "Which legislation requires superhumans to be governed by law?\n1. Crimea accord\n2. Latervia accord\n3. Sokovia accord",
                      "What Nazi organization infiltrated SHIELD?\n1. Hydra\n2. Sphinx\n3. Chimera",
                      "Who is Thor\'s father?\n1. Korg\n2. Odin\n3. Loki",
                      "Where did Scott Lang work before becoming Ant-Man?\n1. Jamba Juice\n2. Chipotle\n3. Baskin Robbins",
                      "What fast food do the Avengers eat after the Battle of New York?\n1. Shawarma\n2. Falafel\n3. Noodles",
                      "Who was the keeper of the Soul Stone at Vormir?\n1. Hela\n2. Red Skull\n3. The Destroyer"]

    ans_hard_list = ["1",
                     "2",
                     "2",
                     "3",
                     "3",
                     "1",
                     "2",
                     "3",
                     "1",
                     "2"]
    
    

    counter = 2
    while counter >=0:
        question = random.choice(ques_hard_list)
        print(question)
       
       
        
        while True:
            ans = input("<Plese enter '1', '2', '3' only \n>")
            
            try:
                int(ans)
                break
 
            except:
                print("\nYou choose something that is not acceptable. Please choose between '1', '2' and '3'. Choose again.")     
                
        # I made people play the game and asked them if they had a problem with answering the questions again
        # if they input 0 or a number bigger than 4. The people who were Marvel fans did not mind answering these 
        # questions again but the people who had not watched Marvel movies and did not know anything about the 
        # Avengers did not enjoy answering the questions again.
        if ans >= '4' or ans == '0':
            print("\nYou choose a wrong number. \nNow answer these questions again. Be careful while choosing.")
            superman_ques()
        

        elif ans == ans_hard_list[ques_hard_list.index(question)]:
            print("\nYou are right!")
            ans_hard_list.pop(ques_hard_list.index(question))
            ques_hard_list.pop(ques_hard_list.index(question))
            input('\n<Press enter to continue>\n')
            

        elif ans_hard_list.index(ans) != ans_hard_list[ques_hard_list.index(question)]:
            print(f"""
    Wrong answer. Superman kills you. 
            """)        
            ans_hard_list.pop(ques_hard_list.index(question))
            ques_hard_list.pop(ques_hard_list.index(question))
            fail()

        else: 
            print("\nInvalid answer! Please try again")
            superman_ques()
            break

        counter -=1

    superman_win()

def superman_win():
    print("""
    Superman helps the avengers find and defeat Thanos.
    The snap was undone and all the people who had died because of the snap were saved.
    Congratulations!!
    You won!
    You saved the earth!""")
    exit(0)
        
def fail():
    print("\nYou have failed to save the world!")
    print("\nDo you wish to try again?\n1. Yes\n2. No") 
    replay = input("> \n").lower().strip()
    if "1" in replay or "yes" in replay or "y" in replay: 
        print("\nI hope you are able to save the world this time!") 
        round_1()
        
    elif "2" in replay or "no" in replay or "n" in replay: 
        print("\nGoodbye!") 
        exit(0)
        
    else: 
        print("\nInvalid entry. Please try again.\n")
        fail()
    
game_start()




        
                

                
```
